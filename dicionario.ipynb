```python
import pandas as pd
```

```python
df_dicionario = pd.read_csv('br_me_rais_dicionario.csv')
```

```python
df_dicionario.head()
```

```python
df_dicionario.drop(columns= ['cobertura_temporal'])
```

```python
df_mapas = df_dicionario[df_dicionario['id_tabela'] == 'microdados_vinculos']
```

```python
mapeamento = {}
```

```python
for coluna, group_df in df_mapas.groupby('nome_coluna'):
    # Para cada coluna, criar um dicionário interno de {chave: valor}
    mapa_interno = dict(zip(group_df['chave'], group_df['valor']))
    mapeamento[coluna] = mapa_interno
```
