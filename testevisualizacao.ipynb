```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_parquet(r'C:\Users\User\Documents\Residência em Dados - Instituto ECOA PUCRIO\df_rais_tratado.parquet')
```

```python
fig = px.treemap(
    df,
    # path define a hierarquia. Começa com a UF, depois grau de instrução.
    path=[px.Constant("Brasil"), 'sigla_uf', 'grau_instrucao_apos_2005_label'],
    
    # values define o tamanho de cada retângulo.
    values='valor_remuneracao_media', # Pode ser 'quantidade_horas_contratadas', etc.
    
    # color pode ser o próprio valor (tamanho) ou outra métrica, como a média.
    # Aqui, vamos colorir pela *média* da remuneração (o Plotly calcula automaticamente)
    color='valor_remuneracao_media',
    
    # Define como o valor da cor é calculado (média, mediana, soma...)
    color_continuous_scale='fall', # Escolha uma escala de cor
    
    title='Hierarquia da Remuneração Média por UF e Grau de Instrução'
)
```

```python
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
```

```python
fig.show()
```

```python
fig.write_html("meu_grafico_interativo.html", 
               full_html=False, 
               include_plotlyjs='cdn')
```
